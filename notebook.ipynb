## Popularity Recommendation

Logic: If an item is liked by a majority of the user base, then the item is recommended to the user if the user has not interacted with that item

```python
# assessing popular recommendations for user no. 5
# another user can be selected by passing a different number
tprint('LOAD')
popular_rec = engine.popular_recommendations(5)
print('\n')
tprint('POP  RECS')
popular_rec
```

## Similarity Recommendation

Logic: using the Jaccard index as the similarity metric we find if two items are liked by a majority fraction of common users within our total users, the two items can be said to be similar. If the items are similar, we can recommend the item which has not been interacted with by the user to the user

1. Understand what items the user has interacted with
2. Calculate similarity of each item the user has interacted with to the items in the dataset
3. Discover the items that are most similar to items the user has already interacted with
4. Recommend a subset of the items based on similarity

```python
# assessing similarity recommendations for user no. 5
# another user can be selected by passing a different number
tprint('LOAD')
similarity_rec = engine.similarity_recommendations(5)
print('\n')
tprint('SIM RECS')
similarity_rec
```

## Matrix Factorization Recommendation

In this implementation we are indentifiying simple features based off user interaction and ratings. To explore this further, we could using more domain knowledge and dive in to item characteristics, rather than ratings.
